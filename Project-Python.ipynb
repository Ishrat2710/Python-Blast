```python

import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

file_path = "/titanic.csv"
titanic_df = pd.read_csv(file_path)

# Task 1: Find the number of passengers by embarkation point
embarkation_counts = titanic_df['Embarked'].value_counts()
print("Number of passengers by embarkation point:")
print(embarkation_counts)

#Task-2:# Task 2: Replace missing values in age with the mean age

titanic_df["Age"] = titanic_df["Age"].replace("", pd.NA)
df_filled = titanic_df.fillna({"Name": "Unknown", "Age": titanic_df["Age"].mean()})
print("\nAfter Filling Missing Data:\n", df_filled)

# Task 3: Analyze the fare distribution using a violin plot by survival status
plt.figure(figsize=(10, 6))
sns.violinplot(data=titanic_df, x='Survived', y='Fare', palette='muted')
plt.title('Fare Distribution by Survival Status')
plt.xlabel('Survived (0 = No, 1 = Yes)')
plt.ylabel('Fare')
plt.show()

# Task 4: Count the number of families (passengers with non-zero sibsp or parch)
families_count = titanic_df[(titanic_df['SibSp'] > 0) | (titanic_df['Parch'] > 0)].shape[0]
print(f"\nNumber of families (passengers with non-zero SibSp or Parch): {families_count}")

# Task 5: Find the proportion of male and female passengers in each class
proportions_by_class = titanic_df.groupby(['Pclass', 'Sex']).size() / titanic_df.groupby('Pclass').size()
print("\nProportion of male and female passengers in each class:")
print(proportions_by_class)

# Task 6: Plot a bar chart comparing survival rates across embarkation points
survival_rates_by_embarkation = titanic_df.groupby('Embarked')['Survived'].mean()
plt.figure(figsize=(8, 5))
survival_rates_by_embarkation.plot(kind='bar', color='skyblue', edgecolor='black')
plt.title('Survival Rates by Embarkation Point')
plt.xlabel('Embarkation Point')
plt.ylabel('Survival Rate')
plt.xticks(rotation=0)
plt.show()
```
